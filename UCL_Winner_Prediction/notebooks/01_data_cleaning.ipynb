```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("Libraries imported successfully!")
```

# Load the Dataset

```python

df = pd.read_csv('/Users/nazmul/Desktop/Project/ucl/data/ucl_stats.csv')
print(f"Dataset loaded: {df.shape}")
print(f"\nFirst few rows:")
df.head()
```

```python
# Data Information
print("=" * 70)
print("DATASET INFORMATION")
print("=" * 70)
print(f"\nShape: {df.shape}")
print(f"Columns: {list(df.columns)}")
print(f"\nData Types:")
print(df.dtypes)
print(f"\nMissing Values:")
print(df.isnull().sum())
```

```python
#Check for Duplicates
print("=" * 70)
print("DATA QUALITY CHECK")
print("=" * 70)
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

```

```python

# Check for logical inconsistencies
print("\nChecking logical inconsistencies...")
consistency_check = df['wins'] + df['draws'] + df['losts'] == df['match_played']
inconsistent_matches = (~consistency_check).sum()
print(f"Rows with match inconsistency: {inconsistent_matches}")

gd_check = df['goals_scored'] - df['goals_conceded'] == df['gd']
inconsistent_gd = (~gd_check).sum()
print(f"Rows with goal difference inconsistency: {inconsistent_gd}")
```

```python
#Remove Inconsistencies
initial_rows = len(df)

# Remove records with logical inconsistencies
df_clean = df[consistency_check & gd_check].copy()

final_rows = len(df_clean)
print(f"\n Cleaning Complete:")
print(f"Initial rows: {initial_rows}")
print(f"Final rows: {final_rows}")
print(f"Rows removed: {initial_rows - final_rows}")
print(f"Retention rate: {(final_rows/initial_rows)*100:.2f}%")
```

# Save Cleaned Dataset

```python

df_clean.to_csv('/Users/nazmul/Desktop/Project/ucl/data/ucl_stats_cleaned.csv', index=False)
print(" Cleaned dataset saved as 'ucl_stats_cleaned.csv'")
```
